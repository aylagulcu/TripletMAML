```python
from Triplets import *
from torchvision import transforms
import os
os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"
```

```python
tripletomniglot = TripletMiniImageNet( root="./data", transform = transforms.Compose([transforms.ToTensor()]))
```

```python
import matplotlib.pyplot as plt

def show_triplets(img1, img2, img3, img4, label1,label2,label3, label4):
    """Display image for testing"""
    fig = plt.figure(figsize=(10, 10))
    ax1 = fig.add_subplot(141)
    ax2 = fig.add_subplot(142)
    ax3 = fig.add_subplot(143)
    ax4 = fig.add_subplot(144)
    ax1.imshow(img1.permute(1, 2, 0).int())
    ax2.imshow(img2.permute(1, 2, 0).int())
    ax3.imshow(img3.permute(1, 2, 0).int())
    ax4.imshow(img4.permute(1, 2, 0).int())
    ax1.title.set_text(label1)
    ax2.title.set_text(label2)
    ax3.title.set_text(label3)
    ax4.title.set_text(label4)
    plt.show()
    
def show_helper(tripletx,triplety):
    for i in range(len(tripletx)):
        show_triplets(*tripletx[i],*triplety[i])
```

```python
a = tripletomniglot.sample("train")
```

```python
show_helper(a[0],a[1])
```
